#Dataset Description
 The dataset on heart diseases is gathered from Kaggle. It is among the most great sources for datasets being used in machine learning research (Banachewicz and Massaron, 2022). This dataset includes information on the medical records of patients and is helpful in studying heart disease. The dataset has several important features and attributes. These include:
Age: The age of the patient.
Sex: The gender of the patient: 1= male, 0 = female.
Chest pain type: Type of chest pain experienced from 1 to 4.
Resting blood pressure: The patient's resting blood pressure.
Cholesterol: Level of the patient's cholesterol.
Fasting blood sugar: refers to a binary variable of whether the patient's fasting blood sugar > 120 mg/dl.
Resting ECG: patient electrocardiogram results (0 to 2 values).
Maximum Heart Rate: Highest achieved rate of heart-beats by the patient.
Exercise-induced angina: The patient having angina with exercise, 1 = yes, 0 = no.
Old peak: The amount of ST depression at peak exercise relative to rest.
ST Slope: slope of the peak exercise ST segment, values 1 to 3. Target: Presence of heart disease (1 = yes, 0 = no).
One of the primary reasons for choosing this dataset is its wide application in the medical and machine learning communities, it would be a good benchmark for comparisons. The set contains numerical and categorical features, allowing various techniques to demonstrate data preprocessing and integration. The fact that it focuses on heart disease is relevant and important in itself, since it is one of the leading causes of death worldwide. Such a subject makes the analysis not only technically interesting but socially significant too.

# Implementation of Hadoop MapReduce
In this section, the analysis explains how to simulate a MapReduce job using Python. The analysis have the entire process of simulation organized into three significant steps: a map step, a shuffle and sort step, and a reduce step.
# Map Step: Creation of Key-Value Pairs
In the map step, the analysis creates key-value pairs from the dataset. This can be a single gender and cholesterol level, where the former is the key and the latter the value. The step structures data for further processing by associating with each measure of cholesterol a gender.
# Shuffle and Sort Step: Group by Key
The analysis performs the shuffle and sort step of grouping all the key-value pairs by keyin our case, gender. So, the sorting is done such that all cholesterol readings for every gender are bunched together now. This organization would be necessary for the reduction step that follows.
# Reduce Step: Aggregation and Final Output
The analysis executes the reduce step in which the analysis aggregates the values of cholesterol corresponding to each gender by taking the average level of cholesterol. The result from the aggregation is taken as a final output to provide an average cholesterol level for males and females. Results from the MapReduce implementation provide the average cholesterol by gender. It could be that females have an average cholesterol level of 0.374897, while males have 0.115892. As noted, the simulated implementation conducted the core steps involved in any MapReduce job: mapping, shuffling, sorting, and reducing. Python can be effectively used to simulate all of these and learn insights into the data processing power of Hadoop MapReduce.

# Logistic Regression for Prediction of Heart Diseases using Spark
It predicts heart disease based on relevant features extracted from the dataset. The features considered at this point include age, sex, type of chest pain, resting blood pressure, cholesterol level, fasting blood sugar, resting electrocardiographic results, maximum heart rate during exercise, exercise-induced angina, old peak, and ST slope. Having seen these various features, what remains is to put these features in one vector and use this as input for the logistic regression model which is a crucial step in preparing data for machine learning.
# Model Training and Evaluation
The analysis then splits the dataset into a train set and a test set to establish the model's performance. First would be the training set to train a logistic regression model, then depicts the use of the test set to see how accurate the logistic regression model is. Due to the input features, the logistic regression model learns how to predict the Presence of Heart Disease variable. The analysis then evaluates the performance using test sets for this trained model. The accuracy is computed, which tells one how good the model was at predicting heart disease. Results from Spark Implementation includes calculating the averages of cholesterol levels for males and females. Moreover, logistic regression achieves an accuracy of 81.84% on test data. This accuracy tells that heart disease can be efficiently predicted with modeled features. It enables us to carry out effective data processing and analysis, proving the benefits over traditional MapReduce. In-memory processing enables faster computations and spark becomes a strong tool in large-scale data analyses and other tasks of machine learning. Using Spark helps achieve not only fast processing of data but also easy implementation in comparison with the already mentioned traditional MapReduce (Shaikh et al., 2019). It runs complex data processing tasks, showing the efficiency of Spark, performance, and usability.


```python
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.feature import VectorAssembler

feature_columns = ['age', 'sex', 'chest pain type', 'resting bp s', 'cholesterol',
                   'fasting blood sugar', 'resting ecg', 'max heart rate',
                   'exercise angina', 'oldpeak', 'ST slope']

assembler = VectorAssembler(inputCols=feature_columns, outputCol="features")
data = assembler.transform(heart_data_spark)

final_data = data.select("features", col("target").alias("label"))

train_data, test_data = final_data.randomSplit([0.7, 0.3], seed=42)

lr = LogisticRegression(featuresCol="features", labelCol="label")

lr_model = lr.fit(train_data)

test_results = lr_model.evaluate(test_data)

print("Test Accuracy:", test_results.accuracy)
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="accuracy")
accuracy = evaluator.evaluate(test_results.predictions)
print(f"Accuracy: {accuracy}")

evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="weightedPrecision")
precision = evaluator.evaluate(test_results.predictions)
print(f"Precision: {precision}")

evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="weightedRecall")
recall = evaluator.evaluate(test_results.predictions)
print(f"Recall: {recall}")

evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="f1")
f1 = evaluator.evaluate(test_results.predictions)
print(f"F1 Score: {f1}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

predictions_pd = test_results.predictions.select("label", "prediction").toPandas()

confusion_matrix = pd.crosstab(predictions_pd['label'], predictions_pd['prediction'], rownames=['Actual'], colnames=['Predicted'])

plt.figure(figsize=(10, 7))
sns.heatmap(confusion_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.show()
```

# Performance Comparison
The following section compares Hadoop MapReduce and Apache Spark across several execution parameters or criteria, most especially on execution time, resource utilization, scalability, fault tolerance, and efficiency in data shuffling and sorting.
# Execution Time Measurement of MapReduce and Spark Tasks
The analysis measures the execution time for both MapReduce and Spark by running similar tasks on each framework. The analysis compares them in this exercise by calculating the average cholesterol levels by gender. The analysis records the start and end times of each task and compute the total execution time.
# Results of the Execution Time Measurements
Results show that Apache Spark runs the job much faster than Hadoop MapReduce. This can be attributed to the data handling efficiency brought about by the in-memory processing capability of Spark, hence shorter execution times.
# Resource Utilization Comparison
Monitoring tools, like Ganglia for Hadoop and Spark's built-in web UI, track resource usage in terms of CPU, memory, and I/O. The analysis profiles the details of resource usage during the execution of tasks. The experimental results show that the resource consumption by Spark is relatively low compared to Hadoop MapReduce. This is because in-memory processing used in Spark reduces the extent of disk I/O and hence results in lower usage of memory and CPU. More importantly, Hadoop MapReduce has very intensive disk operations, raising overall resource consumption.
# Scalability Testing
The analysis tests scalability by running tasks on datasets of an increasing size. The analysis starts with smaller datasets and gradually increase the size to examine the way each handled the growing data. The results indicate that Spark scales much better compared to Hadoop MapReduce. Specifically, Spark increases the execution time and resource usage slightly along with the data size, whereas Hadoop MapReduce increases its execution time exponentially times and its resource usage tremendously.
# Efficiency in Data Shuffling and Sorting
The analysis further measures the efficiency of data shuffling and sorting by timing these operations during task execution. Spark does these more efficiently because its in-memory computation model eliminates the significant overhead associated with Hadoop MapReduce because of its disk-based operations. Community activity and support of both Hadoop MapReduce and Apache Spark have large, active communities, together with rich support (Chos, 2019). However, the growing popularity has shifted Spark even further in terms of community members and response resources across documentation, tutorials, and forums, making it far easier to get help and support. In-memory processing makes Spark much better in handling big data processing tasks due to its advanced features.
